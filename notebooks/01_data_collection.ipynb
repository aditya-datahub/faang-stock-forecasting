# 📥 Notebook 1 — Data Collection
**Project:** FAANG Stock Price Forecasting  
**Stocks:** META, AAPL, AMZN, NFLX, GOOGL  
**Goal:** Download 5 years of historical stock data and save it for further analysis.

## Step 1 — Install & Import Libraries

```python
import yfinance as yf
import pandas as pd
import os

print('✅ Libraries imported successfully')
```

## Step 2 — Define Stocks & Date Range

```python
TICKERS    = ["META", "AAPL", "AMZN", "NFLX", "GOOGL"]
START_DATE = "2019-01-01"
END_DATE   = "2024-12-31"

print("Stocks :", TICKERS)
print("From   :", START_DATE)
print("To     :", END_DATE)
```

## Step 3 — Download Data for All 5 Stocks

```python
stock_data = {}

for ticker in TICKERS:
    print(f"Downloading {ticker}...", end=" ")
    df = yf.download(ticker, start=START_DATE, end=END_DATE, progress=False)

    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    df["Ticker"] = ticker
    stock_data[ticker] = df
    print(f"✅  {len(df)} rows")

print("\nAll stocks downloaded!")
```

## Step 4 — Preview Each Stock

```python
for ticker, df in stock_data.items():
    print(f"── {ticker} ──")
    print(f"   Date range : {df.index.min().date()} → {df.index.max().date()}")
    print(f"   Rows       : {len(df)}")
    print(f"   Columns    : {df.columns.tolist()}")
    print()
```

## Step 5 — Combined Close Price Table

```python
close_df = pd.concat(
    {ticker: df["Close"] for ticker, df in stock_data.items()},
    axis=1
)

print(f"Shape: {close_df.shape}")
close_df.head(10)
```

```python
close_df.describe().round(2)
```

## Step 6 — Check & Fix Missing Values

```python
print("Missing values per stock:")
print(close_df.isnull().sum())

close_df = close_df.ffill()

print("\nAfter forward fill:")
print(close_df.isnull().sum())
```

## Step 7 — Save Data to CSV

```python
os.makedirs("data", exist_ok=True)

close_df.to_csv("data/FAANG_close_prices.csv")
print("✅ Saved: data/FAANG_close_prices.csv")

for ticker, df in stock_data.items():
    path = f"data/{ticker}_raw.csv"
    df.to_csv(path)
    print(f"✅ Saved: {path}")

print(f"\nTotal files saved: {len(TICKERS) + 1}")
```
